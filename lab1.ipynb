```python
#Assignment 3
#part a
import simpleaudio as sa
wav_obj=sa.WaveObject.from_wave_file('train32.wav')
fs1=wav_obj.sample_rate
channel1=wav_obj.num_channels

print('Number of  train channels=',channel1)


wav_obj=sa.WaveObject.from_wave_file('tuba11.wav')
fs2=wav_obj.sample_rate
channel2=wav_obj.num_channels

print('Number of tuba channels=',channel2)

#partb
from scipy.io import wavfile as wav
fs1,data1=wav.read('train32.wav')
y1=data1



fs2,data2=wav.read('tuba11.wav')
y2=data2
len2,ch2=y2.shape

#partb
play_obj =sa .play_buffer(y1, channel1,2, fs1)
play_obj.wait_done()
play_obj =sa .play_buffer(y2, channel2,2, fs2)
play_obj.wait_done()
train_swap = sa.play_buffer(y1, 1, 2, fs2) 
train_swap.wait_done()
tuba_swap = sa.play_buffer(y2, 2, 2, fs1)
tuba_swap.wait_done()



# Part C
print('Train whistle has :sampling rate',fs1,',#of samples',len(y1),',type',y1.dtype)
print('Tuba has:sampling rate',fs2,',#of samples',len2,',#of channels',ch2,',type',y2.dtype)

y2 = y2[:, 0]
data1 = len(y1)
data2 = len(y2)
if data1 > data2:
    y3 = y1[0 : data2]
    y4 = 0.5 * y3 + 0.5 * y2
    y4 = np.int16(y4)
else:
    y3 = y2[0 : data1]
    y4 = 0.5 * y3 + 0.5 * y1
    y4 = np.int16(y4)
play_obj = sa.play_buffer(y4, channel1, 2, fs1)
play_obj.wait_done()
outfile = 'y4.wav'
wav.write(outfile,fs1,y4.astype('int16'))

# Part D
pause = np.zeros(int(4 * fs1), dtype = np.int16)
y5 = np.concatenate([y1, pause, y2])
play_obj = sa.play_buffer(y5, channel1, 2, fs1)
play_obj.wait_done()
outfile = 'y5.wav'
wav.write(outfile,fs1,y5.astype('int16'))
```

```python
# Assignment 4
%matplotlib notebook
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

# Part A
fs=5

t= np.arange(-2,4.1, (1/fs))



# Part B

w=abs(t)
x=2-t
y=-0.5*(t**2)

fig1=plt.figure(1)

plt.plot(t,w,label='w(t)')
plt.plot(t,x,label='x(t)')
plt.plot(t,y,label='y(t)')
plt.xlabel('t')
plt.suptitle("comparison plot")
plt.legend()
plt.show

plt.grid()







```

```python
# Assignment 5 -- Plotting Sound Files using Subplots

# Part A
from scipy.io import wavfile as wav
fs1, y1 = wav.read('train32.wav')
fs2, y2 = wav.read('tuba11.wav')
t1 = np.arange(0, len(y1), 1)
t1 = t1 / fs1 * 1000
t2 = np.arange(0, len(y2), 1)
t2 = t2 / fs2 * 1000

# Part B
fig2 = plt.figure(2)
fig2.subplots_adjust(hspace = 0.5, wspace = 0.2)

plt.subplot(1, 2, 1)
plt.plot(t1, y1, color = 'red')
plt.title('y1')
plt.ylim(-20000, 20000)
plt.xlim(0, 2500)
plt.ylabel('Amplitude')
plt.xlabel('Time(ms)')

plt.subplot(1, 2, 2)
plt.plot(t2, y2, color = 'yellow')
plt.title('y2')
plt.ylim(-20000, 20000)
plt.xlim(0, 2500)
plt.ylabel('Amplitude')
plt.xlabel('Time(ms)')
plt.show()

```
